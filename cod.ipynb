```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats import kurtosis
from scipy.stats import variation
import matplotlib.dates as mdates
```

```python
data  = pd.read_csv('data.csv')
```

```python
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=90))
#plt.plot(np.flip(data.date),np.flip(data.sold))
plt.plot(np.flip(data.date),(np.convolve(np.flip(data.hidro), np.ones(30), 'same') / 30))
plt.plot(np.flip(data.date),(np.convolve(np.flip(data.nuclear), np.ones(30), 'same') / 30))
plt.gcf().autofmt_xdate()
plt.xlabel("Date")
plt.ylabel("MW")
plt.tight_layout()
```

```python
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=90))
plt.plot(np.flip(data.date),np.flip(data.sold))
plt.plot(np.flip(data.date),(np.convolve(np.flip(data.sold), np.ones(30), 'same') / 30))
plt.gcf().autofmt_xdate()
plt.xlabel("Date")
plt.ylabel("MW")
plt.tight_layout()
plt.savefig('plot_sold.png')
```

```python
print(data.carbune.mean())
print(np.std(data.carbune))
print(skew(data.carbune))
print(kurtosis(data.carbune))
print(variation(data.carbune))
print(data.carbune.min())
print(data.carbune.max())

plt.hist(data.carbune, bins ='auto')

plt.xlabel("MW")
plt.ylabel("Frequency")
plt.savefig('hist_coal.png')
```
